```python
from src.augment.maze.point_maze_aug_function import PointMazeTrajectoryAugmentationFunction, PointMazeGuidedTrajectoryAugmentationFunction
import gym
import h5py
import gzip
from src.generate.utils import npify

trajectory = {
    'observations': observed_dataset['observations'][start_timestamp:end_timestamp+1],
    'actions': observed_dataset['actions'][start_timestamp:end_timestamp+1],
    'next_observations': observed_dataset['next_observations'][start_timestamp:end_timestamp+1],
    'rewards': observed_dataset['rewards'][start_timestamp:end_timestamp+1],
    'terminals': observed_dataset['terminals'][start_timestamp:end_timestamp+1],
}

env = gym.make('maze2d-umaze-v0')
env.reset()
f1 = PointMazeTrajectoryAugmentationFunction(env=env)


print("random augmented trajectory")
augmented_trajectory = f1.augment_trajectory(trajectory)

# env = gym.make('maze2d-umaze-v0')
# env.reset()
print("guided augmented trajectory")
f2 = PointMazeGuidedTrajectoryAugmentationFunction(env=env)
guided_augmented_trajectory = f2.augment_trajectory(trajectory)

trajectory_dataset = h5py.File("trajectory.hdf5", 'w')
npify(trajectory)
for k in trajectory:
    trajectory_dataset.create_dataset(k, data=trajectory[k], compression='gzip')
trajectory_dataset.close()

augmented_trajectory_dataset = h5py.File("random_augmented_trajectory.hdf5", 'w')
npify(augmented_trajectory)
for k in augmented_trajectory:
    augmented_trajectory_dataset.create_dataset(k, data=augmented_trajectory[k], compression='gzip')
augmented_trajectory_dataset.close()


guided_augmented_trajectory_dataset = h5py.File("guided_augmented_trajectory.hdf5", 'w')
npify(guided_augmented_trajectory)
for k in guided_augmented_trajectory:
    guided_augmented_trajectory_dataset.create_dataset(k, data=guided_augmented_trajectory[k], compression='gzip')
guided_augmented_trajectory_dataset.close()

```
